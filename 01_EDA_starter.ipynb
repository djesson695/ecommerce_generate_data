# 1.3 - Analyse des Distributions

**TODO :** Visualisez les distributions principales


1. Distribution des âges des utilisateurs

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(users_df["age"], bins=50, color='skyblue', edgecolor='black')
plt.title("Distribution des âges des utilisateurs", fontsize=16, fontweight='bold')
plt.xlabel("Age ?", fontsize=12)
plt.ylabel("Nombre de personnes ageées", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

2. Distribution des prix des produits


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(products_df["price"], bins=50, color='skyblue', edgecolor='black')
plt.title("Distribution des prix des produits", fontsize=16, fontweight='bold')
plt.xlabel("Prix (€)", fontsize=12)
plt.ylabel("Nombre de produits", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

3. Distribution des types d'interactions

```python
import matplotlib.pyplot as plt

interaction_counts = interactions_df["interaction_type"].value_counts()

plt.figure(figsize=(8, 5))
interaction_counts.plot(kind='bar', color='lightcoral', edgecolor='black')

plt.title("Distribution des types d'interactions", fontsize=16, fontweight='bold')
plt.xlabel("Type d'interaction", fontsize=12)
plt.ylabel("Nombre d'occurrences", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

4. Top 10 des catégories de produits

```python
import matplotlib.pyplot as plt

top_categories = (
    products_df["category"]
    .value_counts()
    .head(10)
)

plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title("Top 10 des catégories de produits", fontsize=16, fontweight='bold')
plt.xlabel("Catégories", fontsize=12)
plt.ylabel("Nombre de produits", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Compte combien d'interactions chaque utilisateur a eues
interactions_per_user = interactions_df["user_id"].value_counts()

# Affiche un aperçu
print(interactions_per_user.head())

```
